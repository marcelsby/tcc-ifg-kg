````python
def fix_wrong_ufs(row):
    cidade_nascimento = row['cidade_nascimento'].lower()

    if cidade_nascimento in ['arinos', 'indianópolis']:
        row['uf_nascimento'] = 'MG'
    elif cidade_nascimento == 'porto nacional':
        row['uf_nascimento'] = 'TO'
    elif cidade_nascimento == 'brasilia':
        row['uf_nascimento'] = 'DF'

    return row
````

````python
curriculos.loc[curriculos_uf_incorreta.index] = curriculos.loc[curriculos_uf_incorreta.index].apply(fix_wrong_ufs, axis=1)

curriculos.loc[curriculos_uf_incorreta.index]
````
